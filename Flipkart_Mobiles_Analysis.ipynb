# Handling missing storage and memory values

```python
df['Storage']=df['Storage'].fillna(df['Storage'].mode()[0])
df['Memory']=df['Memory'].fillna(df['Memory'].mode()[0])
df.isnull().sum()
```

```python
df['Storage'].unique()
```

# Show the unique number

```python
df['Storage'].nunique()
```

# Extracting Numeric Part of Storage

```python
df['Storage'].value_counts()
```

```python
df['Storage_raw']=df['Storage']
```

```python
df['Storage_value']=df['Storage_raw'].str.extract(r'(\d+)').astype(float)
df['Storage_raw']
```

# Detect the MB values

```python
df['is_MB']=df['Storage_raw'].str.contains('MB',case=False)
df['is_MB']
```

# Detect TB values

```python
df['is_TB']=df['Storage_raw'].str.contains('TB',case=False)
df['is_TB']
```

# Detect GB value

```python
df['is_GB']=df['Storage_raw'].str.contains('GB',case=False)
df['is_GB']
```

# converting everything to GB

```python
df['Storage_value']=np.where(df['is_TB'],df['Storage_value']*1024,df['Storage_value'])
df['Storage_raw']=np.where(df['is_MB'],df['Storage_value']/1024,df['Storage_value'])
df['Storage_value']
```

# Extracting Numeric Part of Memory

```python
df['Memory'].value_counts()
```

```python
df['Memory_raw']=df['Memory']
df['Memory_value']=df['Memory_raw'].str.extract(r'(\d+)').astype(float)
df['Memory_raw']
```

# Detect the MB values

```python
df['is_MB']=df['Memory_raw'].astype(str).str.contains('MB',case=False)
df['is_MB']
```

# Detect TB Values

```python
df['is_TB']=df['Memory_raw'].str.contains('TB',case=False)
df['is_TB']
```

# Detect GB values

```python
#df['is_TB']=df['Memory_raw'].str.contains('TB',case=False)
#df['is_TB']
```

#Converting Everything to GB

```python
df['Memory_value']=np.where(df['is_TB'],df['Memory_value']*1024,df['Memory_value'])
df['Memory_value']=np.where(df['is_MB'],df['Memory_value']/1024,df['Memory_value'])
df['Memory_value']
```

Price Distribution

# BARPLOT

# Average Price Per Memory

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Memory_value',y='Selling Price',data=df)
plt.xlabel('Memory',color='red',fontsize=20)
plt.ylabel('Average Price',color='green',fontsize=15)
plt.xticks(rotation=90)
plt.title('Average Price Per Memory',fontname='sans-serif',color='blue')
plt.show()
```

# Best Value Phone

# Discount
